```python
plt.figure(figsize=(16,10))
plt.bar(top_categories.index,top_categories.values)
plt.xlabel("Country name")
plt.title("Highest population top 5")
plt.show()
```

```python
# Select the top 5 countries based on the population in 1970
top_countries_1970 = df.nlargest(5, '1970')

# Melt the DataFrame to reshape it for visualization
melted_df = top_countries_1970.melt(id_vars=['Country Name'], value_vars=['1960', '1970'], var_name='Year', value_name='Population')

# Create a vertical bar graph
plt.figure(figsize=(10, 6))
plt.bar(melted_df['Country Name'], melted_df['Population'], color=melted_df['Year'].apply(lambda x: 'blue' if x == '1960' else 'orange'))
plt.xlabel('Country Name')
plt.ylabel('Population')
plt.title('Comparison of Population for Top 5 Countries in 1960 and 1970')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Melt the DataFrame to reshape it into a more usable form
melted_df = df.melt(id_vars=["Country Name", "Country Code", "Indicator Name", "Indicator Code"],
                    var_name="Year", value_name="Population")

# Convert the 'Year' column to numeric values
melted_df["Year"] = pd.to_numeric(melted_df["Year"], errors="coerce")

# Find the country with the highest population
country_with_highest_population = melted_df.loc[melted_df["Population"].idxmax()]["Country Name"]
highest_population_country_data = melted_df[melted_df["Country Name"] == country_with_highest_population]

# Plot a bar graph for the highest population country
plt.figure(figsize=(10, 8))
plt.bar(highest_population_country_data["Year"], highest_population_country_data["Population"])
plt.xlabel("Year")
plt.ylabel("Population")
plt.title(f"Highest Population Country: {country_with_highest_population}")
plt.xticks(rotation=90)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Calculate the correlation matrix
correlation_matrix = df.iloc[:, 4:].corr()

# Create a heatmap
plt.figure(figsize=(10, 8))
plt.title("Correlation Heatmap")
sns.heatmap(correlation_matrix, cmap="coolwarm", annot=True, fmt=".2f")
plt.show()
```

```python
# Remove rows with non-numeric identifiers (e.g., 'WLD')
numeric_columns = df.columns[4:]  # Assuming population columns start from index 4
df = df[df[numeric_columns].apply(lambda x: pd.to_numeric(x, errors='coerce')).notna().all(axis=1)]

# Select the top 5 countries based on the population in 2022
top_countries_2022 = df.nlargest(5, '2022')

# Melt the DataFrame
melted_df = top_countries_2022.melt(id_vars=['Country Name'], value_vars=df.columns[4:], var_name='Year', value_name='Population')

# Convert Population column to numeric
melted_df['Population'] = pd.to_numeric(melted_df['Population'])

# Create a bar plot
plt.figure(figsize=(10, 6))
sns.barplot(data=melted_df, x='Country Name', y='Population')
plt.xlabel('Country Name')
plt.ylabel('Population')
plt.title('Population of Top 5 Countries (1960-2022)')
plt.legend(title='Year')
plt.tight_layout()
plt.show()


```

```python

selected_year = '2020'
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x=selected_year, bins=10, kde=True)
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.title(f'Population Distribution in {selected_year}')
plt.tight_layout()
plt.show()
```

```python
# Melt the DataFrame
melted_df = df.melt(id_vars=['Country Name'], value_vars=df.columns[1:], var_name='Year', value_name='Population')

# Convert Population column to numeric
melted_df['Population'] = pd.to_numeric(melted_df['Population'], errors='coerce')

# Create a histogram for each year from 1960 to 2022 with different line colors
plt.figure(figsize=(24,16))
for idx, year in enumerate(range(1960, 2023)):
    year_data = melted_df[melted_df['Year'] == str(year)]
    color = sns.color_palette("Set1", n_colors=63)[idx]  # Choose a color from the palette
    sns.histplot(data=year_data, x='Population', bins=10, kde=True, label=year, color=color, alpha=0.5)

plt.xlabel('Population')
plt.ylabel('Frequency')
plt.title('Population Distribution (1960-2022)')
plt.legend(title='Year', fontsize='medium')
plt.tight_layout()
plt.show()
```

```python
# Specify the year you're interested in (2021 in this case)
year = '2021'

# Group by 'Country Name' and select the population for the specified year
grouped_data = df.groupby('Country Name')[year].sum()

# Sort the grouped data in ascending order (low to high)
sorted_data = grouped_data.sort_values()

# Get the bottom 5 countries with the lowest population
bottom_5_countries = sorted_data.head(5)

print(bottom_5_countries)

```

```python
plt.figure(figsize=(10, 6))
bottom_5_countries.plot(kind='bar')
plt.title(f'Bottom 5 Countries with Lowest Population in {year}')
plt.xlabel('Country')
plt.ylabel('Population')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```
